```python
import cv2
import os,argparse
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = "/content/bill.jpeg"
# image_path = "/content/FELV-cat.jpg"
# image_path = "/content/handwritten-documents.webp"
```

```python
pre_processor = "thresh"
# pre_processor = "blur"
```

```python
images = cv2.imread(image_path)
grayscale_image = cv2.cvtColor(images, cv2.COLOR_BGR2GRAY)

# Apply the chosen pre-processing
if pre_processor == "thresh":
    grayscale_image1 = cv2.threshold(grayscale_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
elif pre_processor == "blur":
    grayscale_image1 = cv2.medianBlur(grayscale_image, 3)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(images, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
```

```python
plt.figure(figsize=(10,5))
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.imshow(grayscale_image1, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
filename = "{}.jpg".format(os.getpid())
cv2.imwrite(filename, grayscale_image1)
```

```python
text = pytesseract.image_to_string(filename)
print(text)
```

```python
import json
import torch
from transformers import (AutoTokenizer,
                          MllamaForConditionalGeneration,
                          AutoProcessor,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          pipeline)
```

```python
config_data = json.load(open("config.json"))
HF_TOKEN = config_data["HF_TOKEN"]
```

```python
model_name = "meta-llama/Llama-3.2-11B-Vision"
```

```python
# tokenizer = AutoTokenizer.from_pretrained(model_name, token=HF_TOKEN)
# tokenizer.pad_token = tokenizer.eos_token
```

```python
processor = AutoProcessor.from_pretrained(model_name, token=HF_TOKEN)
chat_template = "<s>[INST] {user_input} [/INST] {model_output}</s>"
processor.chat_template = chat_template
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)
```

```python
model = MllamaForConditionalGeneration.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    quantization_config=bnb_config,
    device_map="auto",
    token=HF_TOKEN
)
```

```python
from PIL import Image
image = Image.open("/content/bill.jpeg")
```

```python
prompt = "<|image|><|begin_of_text|>for the given image, perform object character recognition and extract the text in it."
inputs = processor(image, prompt, return_tensors="pt").to(model.device)

output = model.generate(**inputs, max_new_tokens=128)
response = processor.decode(output[0])
```

```python
import torch
from PIL import Image as PIL_Image
from transformers import MllamaForConditionalGeneration, AutoProcessor, MllamaProcessor

model_id = "meta-llama/Llama-3.2-11B-Vision-Instruct"

model = MllamaForConditionalGeneration.from_pretrained(model_id, device_map="auto", torch_dtype=torch.bfloat16, token=HF_TOKEN)
processor = MllamaProcessor.from_pretrained(model_id, token=HF_TOKEN)

with open("/content/bill.jpeg", "rb") as f:
    raw_image = PIL_Image.open(f).convert("RGB")
conversation = [
    {
    "role": "user",
    "content": [
        {"type": "image"},
        {"type": "text", "text": "Describe this image in two sentences"},
        ],
    },
]

prompt = processor.apply_chat_template(conversation, add_generation_prompt=True,tokenize=False)

inputs = processor(prompt, raw_image, return_tensors="pt").to(model.device)
output = model.generate(**inputs, temperature=0.7, top_p=0.9, max_new_tokens=512)

print("text&image_output: ",processor.decode(output[0])[len(prompt):])
```
